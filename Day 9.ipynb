```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(1,100,20)
```

```python
y1 = x ** 2.0
y2 = x ** 1.5
```

```python
plt.plot(x,y1,"bo-",linewidth= 2,markersize=10,label="First");
plt.plot(x,y2,"gs-",linewidth = 2,markersize=8,label ="Second");
plt.xlabel("X")
plt.ylabel("Y")
plt.axis([-.5,20,-5,200])
```

```python
x = np.logspace(-1,10,50)
y1 = x ** 2.0
y2 = x ** 1.5
```

```python
plt.loglog(x,y1,"bo-",linewidth= 2,markersize=1,label="First");
plt.loglog(x,y2,"gs-",linewidth = 2,markersize=2,label ="Second");
plt.xlabel("X")
plt.ylabel("Y")
```

```python
x = np.logspace(0,1,10)
y = x**2
plt.loglog(x,y,"bo-")
```
